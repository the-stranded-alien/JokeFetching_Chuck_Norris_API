```python
# result = response.decode("utf-8")
# s = json.loads(result)
# print(type(s))
# print(s)
# print(result)
import csv
```

```python
jokeList = []
for url in urlList:
    response = requests.get(url)
    result = response.content
    result = result.decode("utf-8")
    jsonObj = json.loads(result)
    jokeText = jsonObj['value']['joke']
    print(jsonObj['value']['id'])
    jokeList.append(jokeText)
    print(jokeText)
    with open('out.csv','a',newline='') as f:
        new_data = [[jsonObj['value']['id'],jsonObj['value']['joke']]]
        writer = csv.writer(f)
        writer.writerows(new_data)
    
```

```python
commaRemovedJokeList = []
for joke in jokeList:
    joke = joke.replace(",","")
    commaRemovedJokeList.append(joke)
```

```python
outputFile = open("output.csv",'w')
```
